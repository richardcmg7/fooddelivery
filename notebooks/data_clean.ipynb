## Tabla pivote con Base de datos completa

```python
#df = df[:15]
```

```python
# Convertimos a segundos la salida

def cook_time_to_seconds(time):
    minutes = int(time[0:2])
    seconds = int(time[3:5])
    total_seconds = minutes * 60 + seconds 
    return total_seconds
```

```python
# Dummies para productos
print('Creando dummies para productos')
df_pivot = df.pivot_table(index='order_id', columns='product', values='quantity', aggfunc='sum')
df_pivot = df_pivot.fillna(0)
#df_pivot

# Dummies para el nombre de la tienda
print('Creando dummies para la tienda')
df_pivot_ = df.pivot_table(index='order_id', columns='store_name', values='quantity', aggfunc='min')
df_pivot_ = df_pivot_.fillna(0)
#df_pivot_

# Concatenamos las tablas pivote
print('Combinando dummies')
df_X = pd.merge(df_pivot_, df_pivot, on='order_id', how='left')
print(f'La el numero de ordenes es {df_X.shape[0]} y de columnas {df_X.shape[1]}  ')
print()

# Agrupamos por order_id el dataframe inicial
print('organizando detos objetivo')
print('   * agrupando por orden')
df_ = df.drop_duplicates(subset=['order_id'])
# df_15_.reset_index(drop=True)
print('   * seleccionando variables')
df_y = df_[['order_id', 'cook_time']]

# Creando la columnasegundos
print('   * Creando la columna de segundos')
df_y['cook_time_seconds'] = df_y['cook_time'].apply(lambda X: cook_time_to_seconds(str(X)))
print(f'La el numero de resgistros es {df_X.shape[0]}')
print()
# Combinando entradas y salidas
print('**********')
print('Combinando X y y')
df_final = pd.merge(df_X, df_y, on='order_id', how='left')
#df_final = df_final.drop(['order_id'], axis=1)
df_final
```

```python
df_final.to_csv('df_final.csv')
```
