```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Data Wrangling

```python
data = pd.read_csv('./datasets/df_brunch_munch_medellin_final.csv')
```

## Exploratory Data

```python
plt.figure(figsize=(8, 6))
plt.style.use('fivethirtyeight')
radiation_vs_month = data.loc[:, ['daily_time', 'MonthPS']].groupby('MonthPS').sum()
radiation_vs_month.plot()
plt.xticks(data['MonthPS'], rotation=30)
plt.title('Total cook time by Month')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.style.use('fivethirtyeight')
radiation_vs_month = data.loc[:, ['cook_time_minutes', 'DayPS']].groupby('DayPS').sum()
radiation_vs_month.plot()
plt.xticks(data['DayPS'], rotation=30)
plt.title('Total cook time by Day')
plt.show()
```

### Tiempo promedio de ordenes por fecha

```python
f, ax = plt.subplots(figsize=(20, 5))
sns.lineplot(data= data, x="Date", y="daily_time")

rad_avg= data['daily_time'].mean()
ax.axhline(rad_avg, color='green', linestyle='--', label=f"Mean - {rad_avg:.2f}")

ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth=0.5, color='white',)
ax.grid(which='minor', linestyle=':', linewidth=0.5, color='white', alpha=0.7)

plt.title('Daily Cook Time')
plt.show()
ax.legend()
```

### Tiempo promedio por numero de productos

```python
f, ax = plt.subplots(figsize=(20, 5))
sns.lineplot(data= data, x="Date", y="daily_products_mean")

rad_avg= data['daily_products_mean'].mean()
ax.axhline(rad_avg, color='green', linestyle='--', label=f"Mean - {rad_avg:.2f}")

ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth=0.5, color='white',)
ax.grid(which='minor', linestyle=':', linewidth=0.5, color='white', alpha=0.7)

plt.title('Daily Products Mean')
plt.show()
ax.legend()
```

```python
f, ax = plt.subplots(figsize=(20, 5))
sns.lineplot(data= data, x="Date", y="daily_products_mean")
sns.lineplot(data= data, x="Date", y="daily_time")

rad_avg= data['daily_time'].mean()
ax.axhline(rad_avg, color='green', linestyle='--', label=f"Mean - {rad_avg:.2f}")  

rad_avg= data['daily_products_mean'].mean()
ax.axhline(rad_avg, color='green', linestyle='--', label=f"Mean - {rad_avg:.2f}")

ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth=0.5, color='white',)
ax.grid(which='minor', linestyle=':', linewidth=0.5, color='white', alpha=0.7)

plt.title('Daily Cook Time')
plt.show()
ax.legend()
```

```python
# feature correlation
plt.figure(figsize=(10, 8))
corr= data.corr(method = "pearson")
corr = corr[corr >= 0.5 ]
# plt.matshow(corr)
# plt.show()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap="coolwarm", square=True, annot=True, fmt=".1f")
plt.show()


```

```python
# Adfuller test on target feature

from statsmodels.tsa.stattools import adfuller

cook_time = data['daily_time'].values
result = adfuller(cook_time)

print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value)) # if p-value <= 0.05 the data is stationary, else its non stationary

```
